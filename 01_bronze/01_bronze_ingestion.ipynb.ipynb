```python
file_path = "/Workspace/Food Delivery/synthetic_fooddelivery_dataset.csv"

df = (spark.read
      .format("csv")
      .option("header", "true")
      .option("inferSchema", "true")
      .load(file_path))

target_table = 'foodDelivery.staging.orders_synthetic'

(df.write
 .format('delta')
 .mode('overwrite')
 .saveAsTable(target_table));

dbutils.notebook.exit("SUCCESS")
```
